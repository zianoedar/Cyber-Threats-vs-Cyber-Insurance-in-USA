# Moderen Graph

````python
import matplotlib.pyplot as plt
import numpy as np
import os

# Data
providers = ['Chubb', 'AIG', 'Allianz', 'Zurich', 'AXA XL', 'Hiscox']
first_party = [4, 4, 4, 4, 2, 4]
third_party = [2, 3, 1, 3, 0, 0]
additional = [1, 2, 1, 1, 2, 4]

# Calculate total coverage per provider
total_coverage = np.array(first_party) + np.array(third_party) + np.array(additional)

# Set up bar positions
x = np.arange(len(providers))
bar_width = 0.25

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Background total coverage bars (light gray)
ax.bar(x, total_coverage, width=0.8, color='#E0E0E0', zorder=0, label='Total Coverage')

# Foreground grouped bars with transparency
bars1 = ax.bar(x - bar_width, first_party, bar_width, label='First-Party',
               color='#8cc5e3', edgecolor='none', alpha=0.9, zorder=2)
bars2 = ax.bar(x, third_party, bar_width, label='Third-Party',
               color='#ffbb6f', edgecolor='none', alpha=0.9, zorder=2)
bars3 = ax.bar(x + bar_width, additional, bar_width, label='Additional',
               color='#9fc8c8', edgecolor='none', alpha=0.9, zorder=2)

# Labels & title
ax.set_xlabel('Cyber Insurance Providers', fontsize=14, labelpad=10)
ax.set_ylabel('Number of Coverage Types', fontsize=14, labelpad=10)
# ax.set_title('Cyber Insurance Coverage Breakdown by Provider', fontsize=16, fontweight='bold', pad=15)
ax.set_xticks(x)
ax.set_xticklabels(providers, fontsize=12)
ax.set_ylim(0, max(total_coverage) + 3)
ax.grid(axis='y', linestyle='--', alpha=0.4, zorder=1)

# Data labels for each bar
for bars in [bars1, bars2, bars3]:
    for bar in bars:
        height = bar.get_height()
        if height > 0:
            ax.text(bar.get_x() + bar.get_width()/2, height + 0.2, f'{int(height)}',
                    ha='center', va='bottom', fontsize=10, color='black')

# Data labels for total coverage (background bars)
for i, total in enumerate(total_coverage):
    ax.text(i, total + 0.4, f'Total: {int(total)}', ha='center', fontsize=9, color='gray')

# Legend
ax.legend(fontsize=10, loc='upper center', bbox_to_anchor=(0.5, -0.12), ncol=4, frameon=False)

# Remove spines for a cleaner look
for spine in ['top', 'right']:
    ax.spines[spine].set_visible(False)

plt.tight_layout()

# Save figure
output_path = '/content/modern_coverage_chart.png'
plt.savefig(output_path, dpi=300, bbox_inches='tight')
print(f"Image saved to: {output_path}")

# Display the plot
plt.show()
````
